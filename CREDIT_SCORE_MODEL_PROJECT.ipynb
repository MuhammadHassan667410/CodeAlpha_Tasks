# Importing Data And doing Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/sample_data/Cleaned_CCdata.csv')
```

```python
data.head()
```

```python
data = data.iloc[: , 1:]
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
x = data.drop(['Credit_Score'], axis=1)
y = data['Credit_Score']
```

## Feature selection by mutual information

```python

from sklearn.feature_selection import mutual_info_classif

mutual_info = mutual_info_classif(x, y)

mutual_info_series = pd.Series(mutual_info, index=x.columns)
mutual_info_series = mutual_info_series.sort_values(ascending=False)

display(mutual_info_series)
```

```python

low_mutual_info_columns = mutual_info_series[mutual_info_series < 0.01].index.tolist()
x_filtered = x.drop(columns=low_mutual_info_columns)

display(x_filtered.head())
```

## Scaling the data

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_filtered)
```

# Building Classification models without SMOTE

```python
# Train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

## Random Forest

```python
# Train a Random rorest classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
# predict
y_pred = rf.predict(x_test)
```

```python
# Evaluate using classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

## Logistic Regression

```python
# Train , predict and evaluate using logistic regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)

y_pred_lr = lr.predict(x_test)
print(classification_report(y_test, y_pred_lr))
```

## KNN

```python
# Train , predict and evaluate using KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred_knn = knn.predict(x_test)
print(classification_report(y_test, y_pred_knn))
```

## Decision Tree

```python
# Train , predict and evaluate using Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)
print(classification_report(y_test, y_pred_dt))
```

# Building Classification Models with SMOTE

```python
# Do SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_scaled, y)
```

```python

x_train_resampled, x_test_resampled, y_train_resampled, y_test_resampled = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=42)
```

## Random Forest

```python
# Train, predict and evaluate random forest
rf_resampled = RandomForestClassifier(n_estimators=100, random_state=42)
rf_resampled.fit(x_train_resampled, y_train_resampled)

y_pred_rf_resampled = rf_resampled.predict(x_test_resampled)
print(classification_report(y_test_resampled, y_pred_rf_resampled))
```

## Logistic Regression

```python
# Train , test and evaluate using logistic regression
from sklearn.linear_model import LogisticRegression

lr_resampled = LogisticRegression(random_state=42)
lr_resampled.fit(x_train_resampled, y_train_resampled)

y_pred_lr = lr_resampled.predict(x_test_resampled)
print(classification_report(y_test_resampled, y_pred_lr))
```

## KNN

```python
# Train , test and evaluate using KNN
knn_resampled = KNeighborsClassifier(n_neighbors=5)
knn_resampled.fit(x_train_resampled, y_train_resampled)

y_pred_knn_resampled = knn_resampled.predict(x_test_resampled)
print(classification_report(y_test_resampled, y_pred_knn_resampled))
```

## Decision Tree

```python
# Train , test and evaluate using Decision tree
from sklearn.tree import DecisionTreeClassifier
dt_resampled = DecisionTreeClassifier(random_state=42)
dt_resampled.fit(x_train_resampled, y_train_resampled)

y_pred_dt = dt_resampled.predict(x_test_resampled)
print(classification_report(y_test_resampled, y_pred_dt))
```

# Comparing model Perfomances

```python
# Compare model perfomence
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

models = [rf, lr, knn, dt]
models_resampled = [rf_resampled, lr_resampled, knn_resampled, dt_resampled]

for model, model_resampled in zip(models, models_resampled):
    y_pred = model.predict(x_test)
    y_pred_resampled = model_resampled.predict(x_test_resampled)

    print(f"Model: {model.__class__.__name__}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"Precision: {precision_score(y_test, y_pred, average='weighted')}")
    print(f"Recall: {recall_score(y_test, y_pred, average='weighted')}")
    print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted')}")
    print(f"Resampled Accuracy: {accuracy_score(y_test_resampled, y_pred_resampled)}")
    print(f"Resampled Precision: {precision_score(y_test_resampled, y_pred_resampled, average='weighted')}")
    print(f"Resampled Recall: {recall_score(y_test_resampled, y_pred_resampled, average='weighted')}")
    print(f"Resampled F1 Score: {f1_score(y_test_resampled, y_pred_resampled, average='weighted')}")
    print("\n")
```
